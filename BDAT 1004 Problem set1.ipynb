```python
# question 2 c#


using System;
class program
{
    public static void Main(string[] args)
    {
        string inputString = "Supercalifragilisticexpialidocious";
        string substringToCheck = "ice";

        string firstword    = "Supercalifragilisticexpialidocious";
        string secondword   = "Honorificabilitudinitatibus";
        string thirdword    = "Bababadalgharaghtakamminarronnkonn";

        string[] composers = { "Berlioz", "Borodin", "Brain", "Bartok", "Bellini", "Buxtehude", "Bernstein" };
         Array.Sort(composers);



        int letterCount = CountLetters(inputString);
        bool containsSubstring = CheckSubstring(inputString, substringToCheck);
        string longestWord = FindLongestWord(firstword, secondword, thirdword);
        string firstComposer = composers[0];
        string lastComposer = composers[composers.Length -1];

        
        
        Console.WriteLine($"Number of letters in the string: {letterCount}");

        Console.WriteLine($"Does the string contain ice as substring = {containsSubstring}");

        Console.WriteLine($"The longest word is : {longestWord}");

        Console.WriteLine($"First composer :{firstComposer}");
        Console.WriteLine($"Last composer :{lastComposer}");

            }

    static bool CheckSubstring(string input, string substring)
    {
        return input.Contains(substring);
    }

    static string FindLongestWord(params string[] words) 
    { 
        string longestWord = string.Empty;


       foreach (var word in words) 
        
        {
            if (word.Length > longestWord.Length)
            {
                longestWord = word; 
            }

        }
    return longestWord;
    
    }

   static int CountLetters(string input)
    {
        int count = 0;
        foreach (char c in input)
        {
            if(char.IsLetter(c))
            {
                count++;
            }
        }
        return count;
    }

}

#output

Number of letters in the string: 34
Does the string contain ice as substring = True
The longest word is : Supercalifragilisticexpialidocious
First composer :Bartok
Last composer :Buxtehude


 
```

```python
# question 5 c#

class rectangle
{
    public static void Main()
    {
        Console.WriteLine($"inside(1,1,0,0,2,3): {Inside(1, 1, 0, 0, 2, 3)}"); //true
        Console.WriteLine($"inside(-1,-1,0,0,2,3): {Inside(-1, -1, 0, 0, 2, 3)}");//false

        bool isInBothRectangles = Inside(1, 1, 0.3, 0.5, 1.1, 0.7) && Inside(1, 1, 0.5, 0.2, 1.1, 2);
        Console.WriteLine($"Point (1,1) is inside both rectangles : {isInBothRectangles}");
    }

    static bool Inside(double x, double y, double x1, double y1, double x2, double y2)
    {
        return (x > x1 && x < x2) && (y > y1 && y < y2);
    }

}

#output

inside(1,1,0,0,2,3): True
inside(-1,-1,0,0,2,3): False
Point (1,1) is inside both rectangles : False
```

```python
# question 8

def curconv(currency, amount): 

    with open('currencies.txt', 'r') as file: 

        currency_rates = {} 

        for line in file: 

            parts = line.split() 

            code,rate = parts[0], float(parts[1]) 

            currency_rates[code] = rate 

             

    if currency in currency_rates: 

        converted_amount = amount * currency_rates[currency] 

        return converted_amount 

    else: 

        return "Invalid currency code" 

     

result_eur = curconv('Eur', 100) 

result_jpy = curconv('JPY', 100) 

  

print(result_eur) 

print(result_jpy) 
```

```python
# question 10

def frequencies(text): 

    letters = 'abcdefghijklmnopqrstuvwxyz' 

    frequency_list = [0] * 26 

     for char in txt: 

        char_lower = char.lower() 

        if char_lower in letters: 

            index = letters.index(char_lower) 

            frequency_list[index] += 1 

        return frequency_list 

result1 = frequencies('The quick red fox got bored and went home.') 

result2 = frequencies('apple') 

print(result1) 

print(result2) 
```
